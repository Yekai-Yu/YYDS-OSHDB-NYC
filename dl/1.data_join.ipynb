```python
import pandas as pd

file1 = "2014-2020-trip-feature-v3.csv"
file2 = "2014-2020-node-feature-3-583-appendix.csv"
file3 = "2014-2020-way-feature-3-430-appendix.csv"

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)

cols = ["ZoneNumber", "Year", "Month"]
df2[cols] = df2[cols].astype(int)
df3[cols] = df3[cols].astype(int)

print(len(df1), len(df1.columns))
print(len(df2), len(df2.columns))
print(len(df3), len(df3.columns))
```

```python
df1
```

```python
# Drop year 2020

df1 = df1[df1["Year"]!=2020]
#df1 = df1[~df1["ZoneNumber"].isin([56,57,103,104,105])]
df1
```

```python
# Fix N/A cells
df1[(df1["DOAM"].isna()) | (df1["DOPM"].isna()) | (df1["DOMD"].isna()) | (df1["DOEV"].isna())]
```

```python
df1 = df1.fillna(0)
df1[(df1["DOAM"].isna()) | (df1["DOPM"].isna()) | (df1["DOMD"].isna()) | (df1["DOEV"].isna())]
```

```python
# Normalize Trips
df1.iloc[:,4:12] = df1.iloc[:,4:12].apply(lambda x: (x-x.min())/(x.max()-x.min()), axis=0)
df1
```

```python
# Normalize Node and Way
df2.iloc[:,3:] = df2.iloc[:,3:].apply(lambda x: (x-x.min())/(x.max()-x.min()), axis=0)
df2 = df2.dropna(axis='columns')
df2
```

```python
df3.iloc[:,3:] = df3.iloc[:,3:].apply(lambda x:  (x-x.min())/(x.max()-x.min()), axis=0)
df3 = df3.dropna(axis='columns')
df3
```

```python
df = pd.merge(df1, df2, how = "inner", on = ["ZoneNumber", "Year", "Month"])
```

```python
df = pd.merge(df, df3, how = "inner", on = ["ZoneNumber", "Year", "Month"])
df
```

```python
df.to_csv("final-normalized.csv", index = False, mode = "w")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
df_train = pd.read_csv("final_train.csv")
df_test = pd.read_csv("final_test.csv")
```

```python
Y_train = df_train.iloc[:,:8]
X_train = df_train.iloc[:,8:]

Y_test = df_test.iloc[:,:8]
X_test = df_test.iloc[:,8:]
```

```python
Y_test
```

```python
lreg = LinearRegression().fit(X_train, Y_train)
```

```python
lreg.score(X_train, Y_train)
```

```python
print(lreg.score(X_test, Y_test))
print(mean_squared_error(Y_test, lreg.predict(X_test)))
```

```python
# XGBOOST REGRESSION
import xgboost
xreg_1 = GradientBoostingRegressor(random_state=0).fit(X_train, Y_train.iloc[:,0])
xreg_2 = GradientBoostingRegressor(random_state=0).fit(X_train, Y_train.iloc[:,1])
xreg_3 = GradientBoostingRegressor(random_state=0).fit(X_train, Y_train.iloc[:,2])
xreg_4 = GradientBoostingRegressor(random_state=0).fit(X_train, Y_train.iloc[:,3])
xreg_5 = GradientBoostingRegressor(random_state=0).fit(X_train, Y_train.iloc[:,4])
xreg_6 = GradientBoostingRegressor(random_state=0).fit(X_train, Y_train.iloc[:,5])
xreg_7 = GradientBoostingRegressor(random_state=0).fit(X_train, Y_train.iloc[:,6])
xreg_8 = GradientBoostingRegressor(random_state=0).fit(X_train, Y_train.iloc[:,7])
```

```python
LL = len(X_test)
print(LL)
res = np.zeros((LL, 8))
res[:,0] = xreg_1.predict(X_test)
res[:,1] = xreg_2.predict(X_test)
res[:,2] = xreg_3.predict(X_test)
res[:,3] = xreg_4.predict(X_test)
res[:,4] = xreg_5.predict(X_test)
res[:,5] = xreg_6.predict(X_test)
res[:,6] = xreg_7.predict(X_test)
res[:,7] = xreg_8.predict(X_test)
print(res)
```

```python
print(r2_score(Y_test, res))
print(mean_squared_error(Y_test, res))
```
